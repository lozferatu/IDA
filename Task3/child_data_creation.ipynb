```python
import pandas as pd
import numpy as np
import itertools
```

```python
og_df = pd.read_csv('data.csv',
    header=[0,1]
).sort_index()
```

```python
og_df.head()
```

```python
age_group = ['0-1 years', '2-5 years']
info_list = ['Stunting prevalence in children aged < 5 years (%)', 'Underweight prevalence in children aged < 5 years (%)',
             'Wasting prevalence in children aged < 5 years (%)', 'Severe wasting prevalence in children aged < 5 years (%)' ,
             'Overweight prevalence in children aged < 5 years (%)']
```

```python
temp_df = pd.DataFrame(columns = ['country', 'year'] + info_list)
temp_df['country'] = og_df[('Unnamed: 0_level_0', 'Country')]
temp_df['year'] = og_df[('Unnamed: 1_level_0', 'Year')]
temp_df.sort_values(by = ['country', 'year'], inplace = True)
```

```python
temp_df.head()
```

```python
dfs = []
for age in age_group:
    new_df = temp_df.copy()
    new_df['Age_Group'] = age
    dfs.append(new_df)

# concatenate all the new DataFrames into a single DataFrame
child_data = pd.concat(dfs, ignore_index=True)
```

```python
child_data = child_data[['country', 'year', 'Age_Group', 
                         'Stunting prevalence in children aged < 5 years (%)',
                         'Underweight prevalence in children aged < 5 years (%)',
                         'Wasting prevalence in children aged < 5 years (%)',
                         'Severe wasting prevalence in children aged < 5 years (%)',
                         'Overweight prevalence in children aged < 5 years (%)']]
```

```python
child_data.head()
```

```python
def fill_num_children(row, cols):
    age = row['Age_Group']
    country = row['country']
    year = int(row['year'])
    for i, temp_col in enumerate(cols):
        col = (temp_col, age)
        values = og_df[(og_df[('Unnamed: 0_level_0', 'Country')] == country) & 
                   (og_df[('Unnamed: 1_level_0', 'Year')] == year)][col].values[0]
        row[temp_col] = values
    return row

cols_to_fill = ['Stunting prevalence in children aged < 5 years (%)',
                         'Underweight prevalence in children aged < 5 years (%)',
                         'Wasting prevalence in children aged < 5 years (%)',
                         'Severe wasting prevalence in children aged < 5 years (%)',
                         'Overweight prevalence in children aged < 5 years (%)']
child_data = child_data.apply(lambda row: fill_num_children(row, cols_to_fill), axis=1)
```

```python
child_data.head()
```

```python
child_data.dropna(inplace = True)
```

```python
child_data.to_csv('age_malnutrition.csv')
```
